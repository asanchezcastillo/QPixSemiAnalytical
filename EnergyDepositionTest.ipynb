```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
with open('EdepInformation.txt', 'r') as f:
    EdepInformation = f.readlines()

Edep=[]
dx=[]
dEdx=[]
NElectrons=[]
NPhotons=[]
LightYield=[]
PDG=[]

for num in range(0, len(EdepInformation)):
    if("electrons" in EdepInformation[num]):
        EdepInformation[num]=EdepInformation[num].replace("Number of electrons is ",'').replace("\n",'')
        NElectrons.append(float(EdepInformation[num]))
    elif("photons" in EdepInformation[num]):
        EdepInformation[num]=EdepInformation[num].replace("Number of photons is ",'').replace("\n",'')
        NPhotons.append(float(EdepInformation[num]))  
    elif("Light" in EdepInformation[num]):
        EdepInformation[num]=EdepInformation[num].replace("Light yield is ",'').replace("\n",'')
        LightYield.append(float(EdepInformation[num]))  
    elif("dE/dx" in EdepInformation[num]):
        EdepInformation[num]=EdepInformation[num].replace("dE/dx is ",'').replace("\n",'')
        dEdx.append(float(EdepInformation[num])) 
    elif("Energy deposit" in EdepInformation[num]):
        EdepInformation[num]=EdepInformation[num].replace("Energy deposit is",'').replace("\n",'')
        Edep.append(float(EdepInformation[num])) 
    elif("PDG" in EdepInformation[num]):
        EdepInformation[num]=EdepInformation[num].replace("PDG is",'').replace("\n",'')
        PDG.append(float(EdepInformation[num])) 
    else:
        EdepInformation[num]=EdepInformation[num].replace("dx is ",'').replace("\n",'')
        dx.append(float(EdepInformation[num]))    
    

    
```
